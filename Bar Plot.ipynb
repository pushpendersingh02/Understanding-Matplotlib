# BAR PLOT

```python
import matplotlib.pyplot as plt
import pandas as py
import numpy as np
```

```python
x=['Python','C','JAVA','C++','SQL','HTML5','CSS','PHP','Android']
y=[85,45,70,30,90,65,60,25,50]
plt.bar(x,y)
plt.show()
```

#### Adding Labels to the graph

```python
plt.xlabel('LANGUAGES')
plt.ylabel('POPULARITY')
plt.title('Market Trends', fontsize=25)
plt.bar(x,y)
plt.show()
```

#### Modifying the graph

```python
#Changing the width
plt.xlabel('LANGUAGES')
plt.ylabel('POPULARITY')
plt.title('Market Trends', fontsize=20)
plt.bar(x,y, width=0.5)
plt.show()
```

```python
#Changing the color
plt.xlabel('LANGUAGES')
plt.ylabel('POPULARITY')
plt.title('Market Trends', fontsize=20)
plt.bar(x,y,width=0.5,color='Red')
plt.show()
```

```python
# Keeping color by the legend
# Changing the color
plt.xlabel('LANGUAGES')
plt.ylabel('POPULARITY')
plt.title('Market Trends', fontsize=20)
col=['b','y','b','y','r','g','g','g','m']
plt.bar(x,y,width=0.5,color=col)
plt.show()
```

```python
# Changing the color of the bar edges
plt.xlabel('LANGUAGES')
plt.ylabel('POPULARITY')
plt.title('Market Trends', fontsize=20)
col=['b','b','y','y','m','m','g','g','g']
plt.bar(x,y,width=0.5,color=col,edgecolor='red')
plt.show()
```

```python
# Changing the width of the bar edges
plt.xlabel('LANGUAGES')
plt.ylabel('POPULARITY')
plt.title('Market Trends', fontsize=20)
col=['b','b','y','y','m','m','g','g','g']
plt.bar(x,y,width=0.5,color=col,edgecolor='red',linewidth=3.5)
plt.show()
```

```python
# Changing the line style of the bar edges
# We can use line style as "dotted", "dashed", etc.
plt.xlabel('LANGUAGES')
plt.ylabel('POPULARITY')
plt.title('Market Trends', fontsize=20)
col=['b','b','y','y','m','m','g','g','g']
plt.bar(x,y,width=0.5,color=col,edgecolor='red',linewidth=4, linestyle=':')
plt.show()
```

```python
# Changing the brightness of the bars
# Value of alpha varies between 0 and 1
plt.xlabel('LANGUAGES')
plt.ylabel('POPULARITY')
plt.title('Market Trends', fontsize=20)
col=['b','b','y','y','m','m','g','g','g']
plt.bar(x,y,width=0.5,color=col,edgecolor='red',linewidth=3,linestyle='dashed',alpha=0.25)
plt.show()
```

```python
# Changing the alignment of Data Labels
#Aligned to the edge
plt.xlabel('LANGUAGES')
plt.ylabel('POPULARITY')
plt.title('Market Trends', fontsize=20)
col=['b','b','y','y','m','m','g','g','g']
plt.bar(x,y,width=0.5,color=col,align='edge')
plt.show()
```

```python
# Aligned to the center
# By default it is center aligned
plt.xlabel('LANGUAGES')
plt.ylabel('POPULARITY')
plt.title('Market Trends', fontsize=20)
col=['b','b','y','y','m','m','g','g','g']
plt.bar(x,y,width=0.5,color=col,align='center')
plt.show()
```

```python
#Adding the label to the Graph
plt.xlabel('LANGUAGES')
plt.ylabel('POPULARITY')
plt.title('Market Trends', fontsize=20)
plt.bar(x,y,width=0.5,color='blue',alpha=0.65,label='2024')
plt.show()
```

```python
#Adding the legend to the Graph
plt.xlabel('LANGUAGES')
plt.ylabel('PERCENTAGE')
plt.title('Market Trends', fontsize=20)
plt.bar(x,y,width=0.5,color='blue',alpha=0.65,label='2024 Popularity')
plt.legend()
plt.show()
```

### Multiple Bar Graphs

```python
# Bars stacked/overlapped over each other
x=['Python','C','JAVA','C++','SQL','HTML5','CSS','PHP','Android']
y=[85,45,70,30,90,65,60,25,50]
z=[80,40,80,55,70,65,10,30,70]
w=[50,30,50,95,50,95,15,60,15]
plt.xlabel('LANGUAGES')
plt.ylabel('PERCENTAGE')
plt.title('Market Trends', fontsize=20)
plt.bar(x,y,width=0.5,color='blue',alpha=0.65,label='2024 Popularity')
plt.bar(x,z,width=0.5,color='red',alpha=0.60,label='2020 Popularity')
plt.bar(x,w,width=0.5,color='green',alpha=0.65,label='2016 Popularity')
plt.legend()
plt.show()
```

```python
#Making the bars side-by-side using numpy array/list and width.
x=['Python','C','JAVA','C++','SQL','HTML5','CSS','PHP','Android']
y=[85,45,70,30,90,65,60,25,50]
z=[80,40,80,55,70,65,10,30,70]
w=[50,30,50,95,50,95,15,60,15]

width=0.2
np1=np.arange(len(x)) #Passing the numpy array
np2=[j+width for j in np1]  #It will place the second bar 0.2 away from first
np3=[i+width for i in np2]  #It will place the third bar 0.2 away from second
plt.xlabel('LANGUAGES')
plt.ylabel('PERCENTAGE')
plt.title('Market Trends', fontsize=20)

plt.bar(np1,y,width,color='blue',alpha=0.65,label='2024 Popularity')
plt.bar(np2,z,width,color='red',alpha=0.60,label='2020 Popularity')
plt.bar(np3,w,width,color='green',alpha=0.65,label='2016 Popularity')
plt.legend()
plt.show()
```

```python
# Making the bars side-by-side using numpy array/list and width.
# Retaining the x-axis datalabel using "xticks" parameter
x=['Python','C','JAVA','C++','SQL','HTML5','CSS','PHP','Android']
y=[85,45,70,30,90,65,60,25,50]
z=[80,40,80,55,70,65,10,30,70]
w=[50,30,50,95,50,95,15,60,15]

width=0.2
np1=np.arange(len(x)) #Passing the numpy array
np2=[j+width for j in np1]  #It will place the second bar 0.2 away from first
np3=[i+width for i in np2]  #It will place the third bar 0.2 away from second
plt.xlabel('LANGUAGES')
plt.ylabel('PERCENTAGE')
plt.title('Market Trends', fontsize=20)

plt.bar(np1,y,width,color='blue',alpha=0.65,label='2024 Popularity')
plt.bar(np2,z,width,color='red',alpha=0.60,label='2020 Popularity')
plt.bar(np3,w,width,color='green',alpha=0.65,label='2016 Popularity')

plt.xticks(np1+2*width/2,x)  # Values of np1=[0 1 2 3 4 5 6 7 8], width=0.2, Therefore we had to multiply width by 2 to place it in center as we have records for 3 years. 
plt.legend()
plt.show()
```

```python
# Making the bars side-by-side using numpy array/list and width.
# Retaining the x-axis datalabel using "xticks" parameter
# Rotating the x-axis datalabels
x=['Python','C','JAVA','C++','SQL','HTML5','CSS','PHP','Android']
y=[85,45,70,30,90,65,60,25,50]
z=[80,40,80,55,70,65,10,30,70]
w=[50,30,50,95,50,95,15,60,15]

width=0.25
np1=np.arange(len(x)) #Passing the numpy array
np2=[j+width for j in np1]  #It will place the second bar 0.2 away from first
np3=[i+width for i in np2]  #It will place the third bar 0.2 away from second
plt.xlabel('LANGUAGES')
plt.ylabel('PERCENTAGE')
plt.title('Market Trends', fontsize=20)

plt.bar(np1,y,width,color='blue',alpha=0.65,label='2024 Popularity')
plt.bar(np2,z,width,color='red',alpha=0.60,label='2020 Popularity')
plt.bar(np3,w,width,color='green',alpha=0.65,label='2016 Popularity')

plt.xticks(np1+2*width/2,x,rotation=90)  
# Values of np1=[0 1 2 3 4 5 6 7 8], width=0.2, Therefore we had to multiply width by 2 to place it in center as we have records for 3 years. 
plt.legend()
plt.show()
```

### Horizontal Bar Graph

```python
# Example of Horizontal Graph
x=['Apples','Mangoes','Pineapples','Oranges','Grapes']
y=[334324,543355,882333,123987,433330]

width=0.5
plt.barh(x,y, width)
plt.title('Fruit Sales in Mumbai',fontsize=22)
plt.xlabel('Fruits',fontsize=15)
plt.ylabel('Sales',fontsize=15)

plt.show()
```

### Horizontal Bar Graph with all possible parameters.

```python
x=['Apples','Mangoes','Pineapples','Oranges','Grapes']
y=[334324,543355,882333,123987,433330]
z=[899997,345281,598847,833244,200343]
width=0.35

hp1=np.arange(len(x))
hp2=[i+width for i in hp1]

plt.title('Fruit Sales in Mumbai',fontsize=22)
plt.xlabel('SALE',fontsize=15)
plt.ylabel('FRUITS',fontsize=15)

plt.barh(hp1,y, width, label='2024',color='red',edgecolor='black',linewidth=1.5,linestyle='dashed',alpha=0.7)
plt.barh(hp2,z, width, label='2014',color='yellow',edgecolor='black',linewidth=1.5,linestyle='dashed',alpha=0.7)

plt.legend() # To change the placement of legend we can use 'location' parameter.
plt.yticks(hp1+width/2,x)
plt.show()
```
